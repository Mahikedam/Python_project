# 6. Please check if the age dataset is skewed or symmetric. Based on the results perform the transformation.

```python
# Finding the mean() of the age by taking variable a
a = olympicDF.age.mean()
# Printing the mean of the age
print('Mean of the age = ',a)
# Finding the median() of the age by taking variable b
b = olympicDF.age.median()
# Printing the median of the age
print('Medain of the age =',b)
#  Finding the standard deviation of the age by taking variable c
c = olympicDF.age.std()
# Printig the standard deviation of the age
print('std of the age =',c)
```

```python
olympicDF.age.skew()
```

```python
# Skew = 3 * (Mean â€“ Median) / Standard Deviation
# Performing the Skewness of the age 
print('Skewness only for the age column:')
age_skew = 3 * (a-b)/c
# Printing the Skewness of the age
print('Skewness of age = ',age_skew)
print('\n')
print('Therefore the age dataset is skewed but not symmetric')
print('\n')
# Here the Skewness for overall dataset of the numerical columns
print('Skewness for the age column:')
olympicDF.age.agg(['skew']).transpose()
```

-- A positive skewness value in the output indicates an asymmetry in the distribution corresponding to age and the tail is larger towards the right hand side of the distribution

# 3. Write PySpark code to plot the total number of medals in each Olympic Sport/game.

```python
# Groupbying the sports column and using the aggregate(sum()) function and giving the alias as Number of the medals.
olympicdf_sports_medals = sparkDF.groupBy("Sports").agg(sum("total_medal").alias("Number_of_the_Medals"))
olympicdf_sports_medals.show()
```

-- Here the output tells us the number_of_the_medals of the particular games from the sports cloumn

```python
# importing the seaborn library for better visualization purpose
import seaborn as sns
# Assigning the figure size
plt.figure(figsize=(13,7))
# Assigning rotation to the labels
plt.xticks(rotation = '90')
# plotting the bar plot
ax = sns.barplot(x='Sports',y='Number_of_the_Medals',data=get_the_pandas_df)
```

# 4. Sort the result based on the total number of medals.

```python
# Groupbying the sports column and using the aggregate(sum()) function and giving the alias as Number of the medals by sorting with desc.
olympicDF_medal =  sparkDF.groupBy("sports")\
    .agg(sum("total_medal").alias("Total_number_of_Medals"))\
    .sort(desc("Total_number_of_Medals")).show()
```

-- It shows us like descending the values of number_of_the_medals from high to low values in a manner.

# 5. Find the total number of medals won by each country in swimming.

```python
'''Filtering the swimming game from sports, grouping the country and using 
the aggregate(sum()) function and giving the alias as medals_for_swimming by sorting with desc.'''
olympicDF_swimming =sparkDF.filter("sports =='Swimming'")\
    .groupBy("country")\
    .agg(sum("total_medal").alias("Medals_for_Swimming"))\
    .sort(desc("Medals_for_Swimming")).show()
```

-- The output shows the country column and medals_for_swimming in a manner of desc.

# 6. Find the total number of medals won by each country in Skeleton.

```python
'''Filtering the Skeleton game from sports, grouping the country and using 
the aggregate(sum()) function and giving the alias as medals_for_skeleton by sorting with desc.'''
olympicDF_skeleton =sparkDF.filter("sports =='Skeleton'")\
    .groupBy("country")\
    .agg(sum("total_medal").alias("Medals_for_Skeleton"))\
    .sort(desc("Medals_for_Skeleton")).show()
```

-- -- The output shows the country column and medals_for_skeleton in a manner of desc.

# 7. Find the number of medals that the US won yearly.

```python
'''Filtering the united states from country, grouping the year and using 
the aggregate(sum()) function and giving the alias as Yearly_count_of_medals by sorting the year.'''
olympicDF_US_Medals =sparkDF.filter("country =='United States'")\
    .groupBy("year")\
    .agg(sum("total_medal").alias("Yearly_Count_Of_Medals"))\
    .sort(("year")).show()
```

-- -- The output shows the year column and yearly_count_of_Medals in a manner.

# 8. Find the total number of medals won by each country.

```python
# Groupbying the column column and using the aggregate(sum()) function and giving the alias as Countrywise_medal by sorting with desc.
olympicDF_Countrywise_medal =sparkDF.groupBy("country")\
          .agg(sum("total_medal").alias("Countrywise_medal"))\
          .sort(desc("Countrywise_medal")).show()
```

# 9. Who was the oldest athlete in the Olympics.

```python
# Groupbying the name column and using the aggregate(max()) function and giving the alias as age by sorting with desc.
olympicDF_oldest_athlete =sparkDF.groupBy("name")\
          .agg(max("age").alias("Age"))\
          .sort(desc("Age")).show()
```
